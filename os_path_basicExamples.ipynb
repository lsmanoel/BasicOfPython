**os.path.join(path, *paths)** join one or more path components intelligently. The return value is the concatenation of path and any members of *paths with exactly one directory separator (os.sep) following each non-empty part except the last, meaning that the result will only end in a separator if the last part is empty. If a component is an absolute path, all previous components are thrown away and joining continues from the absolute path component.

**os.path.getmtime(path)** Return the time of last modification of path. The return value is a number giving the number of seconds since the epoch (see the time module). Raise os.error if the file does not exist or is inaccessible.

```
folder_path = "gdrive/My Drive/Colab Notebooks/datasets_image"
dataset_folder = "landscapes"
file_name = "landscape_seek_the_wilds_03.jpg"

if os.path.isdir(os.path.join(folder_path, dataset_folder)):
  result_path = os.path.join(folder_path, dataset_folder, file_name)
  print(result_path)
else:
  print("[Erro] File does not exist!")

#===============================================================================  
epoch_secs = os.path.getmtime(result_path)
print("Epoch time of the last modification: {} seconds:".format(epoch_secs))
print("Date and time: {}".format(time.ctime(epoch_secs)))
#===============================================================================  
```

The return value is a number giving the number of seconds since the epoch (see the time module). Raise os.error if the file does not exist or is inaccessible.

```
folder_path = "gdrive/My Drive/Colab Notebooks/datasets_image"
dataset_folder = "landscapes"
file_name = "landscape_seek_the_wilds_03.jpg"

if os.path.isdir(os.path.join(folder_path, dataset_folder)):
  result_path = os.path.join(folder_path, dataset_folder, file_name)
  print(result_path)
else:
  print("[Erro] File does not exist!")

#===============================================================================  
epoch_secs = os.path.getctime(result_path)
print("Epoch time of the last metadata modification: {} seconds:".format(epoch_secs))
print("Date and time: {}".format(time.ctime(epoch_secs)))
#===============================================================================  
```

**os.path.commonprefix(list)** Return the longest path prefix (taken character-by-character) that is a prefix of all paths in list. If list is empty, return the empty string (''). Note that this may return invalid paths because it works a character at a time.

```
folder_path = "gdrive/My Drive/Colab Notebooks/datasets_image"
dataset_folder = "landscapes"

file_name_1 = "landscape_black_suns_zenith_08.jpg"
file_name_2 = "landscape_progenitus_09.jpg"
file_name_3 = "landscape_hinterland_harbor_10.jpg"

path_1 = os.path.join(folder_path, dataset_folder, file_name_1)
path_2 = os.path.join(folder_path, dataset_folder, file_name_2)
path_3 = os.path.join(folder_path, dataset_folder, file_name_3)

file_path_list = [path_1, path_2, path_3]

#===============================================================================  
common_prex_paths = os.path.commonprefix(file_path_list)
print(common_prex_paths)
#===============================================================================  
```

**glob.glob(pathname)** Return a possibly-empty list of path names that match pathname, which must be a string containing a path specification. pathname can be either absolute (like /usr/src/Python-1.5/Makefile) or relative (like ../../Tools/*/*.gif), and can contain shell-style wildcards. Broken symlinks are included in the results (as in the shell).

```
folder_path = "gdrive/My Drive/Colab Notebooks/datasets_image"
dataset_folder = "landscapes"
feed_path = os.path.join(folder_path, dataset_folder)
type_file = "*.jpg"

file_path_list = []
for file_path in glob.glob(os.path.join(feed_path, type_file)):
  file_path_list.append(file_path)
  print(file_path)
  
common_prex_paths = os.path.commonprefix(file_path_list)
print("\nCommon Prefix: ", common_prex_paths)
```

To search files in a directory tree:

```
folder_path = "gdrive/My Drive/Colab Notebooks/datasets_image/**"
type_file = "*.jpg"
search_path = os.path.join(folder_path, type_file)
print("Search Path: {}".format(search_path))

file_path_list = []
n_files_found = 0

#using enumerate() the value of n_file_found is increment for each file found in file_path
for n_files_found, file_path in enumerate(glob.glob(search_path, recursive=True)):
  file_path_list.append(file_path)

print("{} files found!".format(n_files_found))

common_prex_paths = os.path.commonprefix(file_path_list)
print("Common Prefix: ", common_prex_paths)
```

# Splits

** os.path.splitdrive(path)**
Split the pathname path into a pair (drive, tail) where drive is either a drive specification or the empty string. On systems which do not use drive specifications, drive will always be the empty string. In all cases, drive + tail will be the same as path.

```
folder_path = "gdrive/My Drive/Colab Notebooks/datasets_image"
dataset_folder = "landscapes"
file_name = "landscape_seek_the_wilds_03.jpg"

file_path = os.path.join(folder_path, dataset_folder, file_name)

#===============================================================================  
head, tail = os.path.splitdrive(file_path)
#===============================================================================  

print("head: ", head)
print("tail: ", tail)
```

** os.path.splitext(path)** 
Split the pathname path into a pair (root, ext) such that root + ext == path, and ext is empty or begins with a period and contains at most one period. Leading periods on the basename are ignored; splitext('.cshrc') returns ('.cshrc', '').

```
folder_path = "gdrive/My Drive/Colab Notebooks/datasets_image"
dataset_folder = "landscapes"
file_name = "landscape_seek_the_wilds_03.jpg"

file_path = os.path.join(folder_path, dataset_folder, file_name)

#===============================================================================  
root, ext = os.path.splitext(file_path)
#===============================================================================  
print("root: ", root)
print("extension: ", ext)

print('\n')

#===============================================================================  
root, ext = os.path.splitext(file_name)
#===============================================================================  
print("root: ", root)
print("extension: ", ext)
```


**os.path.split(path)** Split the pathname path into a pair, (head, tail) where tail is the last pathname component and head is everything leading up to that. The tail part will never contain a slash; if path ends in a slash, tail will be empty. If there is no slash in path, head will be empty. If path is empty, both head and tail are empty. Trailing slashes are stripped from head unless it is the root (one or more slashes only). In all cases, join(head, tail) returns a path to the same location as path (but the strings may differ). Also see the functions dirname() and basename().


```
folder_path = "gdrive/My Drive/Colab Notebooks/datasets_image"
dataset_folder = "landscapes"
file_name = "landscape_seek_the_wilds_03.jpg"

file_path = os.path.join(folder_path, dataset_folder, file_name)

#===============================================================================  
head, tail = os.path.split(file_path)
#===============================================================================  
print("head: ", head)
print("tail: ", tail)
```

When we call the method .split(string) (**file_name.split(string)**), the string mark the division point of split:

```
file_name = "landscape_seek_the_wilds_03.jpg"
print(file_name)

splitted_file_name = file_name.split('_seek_the_wilds_')
print(splitted_file_name)

splitted_file_name.append(splitted_file_name[len(splitted_file_name)-1].split('.')[1])
print(splitted_file_name)

splitted_file_name[len(splitted_file_name)-2] = splitted_file_name[len(splitted_file_name)-2].split('.')[0]
print(splitted_file_name)
```

```
folder_path = "gdrive/My Drive/Colab Notebooks/datasets_image"
dataset_folder = "landscapes"
file_name = "landscape_seek_the_wilds_03.jpg"

file_path = os.path.join(folder_path, dataset_folder, file_name)
print(file_path)

splitted_file_path = file_path.split('/')
print(splitted_file_path)

splitted_file_name = splitted_file_path[len(splitted_file_path)-1].split('_')
print(splitted_file_name)

splitted_file_name.append(splitted_file_name[len(splitted_file_name)-1].split('.')[1])
print(splitted_file_name)

splitted_file_name[len(splitted_file_name)-2] = splitted_file_name[len(splitted_file_name)-2].split('.')[0]
print(splitted_file_name)
```
